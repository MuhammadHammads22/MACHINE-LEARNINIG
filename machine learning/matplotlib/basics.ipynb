```python
from cProfile import label
from cgitb import grey
from turtle import color


x=[0,1,2,3]
y=[0,4,5,6]
plt.figure(figsize=(5,4),dpi=300)
plt.plot(x,y,color='grey',linewidth=3,marker="*",linestyle="--",markersize=15,markeredgecolor="red")

plt.title("our firts graph",fontdict={'fontname':'comic sans ms','fontsize':35})
plt.xlabel('X Axis',fontdict={'fontname':'comic sans ms','fontsize':35})
plt.ylabel('Y Axis',fontdict={'fontname':'comic sans ms','fontsize':35})
plt.xticks([1,2,3,4])
# plt.yticks([1,3,5,7])


x2=[0,1,2,3,4,5,6,7.5]

plt.plot(x2[:6],np.square(x2[:6]),"r^--")
plt.plot(x2[5:],np.square(x2[5:]),color="blue",linestyle="-.")
# plt.savefig("firstsgraph.png",dpi=300)
plt.legend()
plt.show()
```
